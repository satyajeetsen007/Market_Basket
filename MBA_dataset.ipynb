```python
#Loading neccesary packages
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#Reading Data
df = pd.read_csv('MBA.csv')
print(df.shape)
df.head()
```

```python
#Since there no column names in the dataset, we'll add the columns manually later, and move the first row to the last, so that we 
#don't miss any records 
df.loc[len(df.index)] = df.columns
print(df.shape)
df.head()
```

```python
#### Data Cleaning
df.replace(np.nan,0,inplace=True) #replacing nan with 0
print(df.shape)
df.head()
```

```python
#### Finding the unique items from the df so, it can be used as our column names
items = []
for i in range(len(df)):
    for j in range(len(df.columns)):
        if(df.iloc[i,j]) == 0:
            pass
        else:
            items.append(df.iloc[i,j])
            
print("Total items purchased :"+str(len(items)))
uniq_items = [*set(items)]
print("Total unique items purchased :"+str(len(uniq_items)))
```

```python
#### creating the dataset as per the apriori algorithm
new_df = pd.DataFrame(df,columns=uniq_items)
print(new_df.shape)
new_df.replace(np.nan,0,inplace=True) #replacing nan with 0
```

```python
for i in range(len(new_df)):
    for j in range(len(uniq_items)):
        a = new_df.iloc[i,j]
        if (a == 0):
            pass
        else:
            new_df.loc[i,a] = 1
            new_df.iloc[i,j] = 0
```

```python
#del new_df[1.0]
new_df.to_csv('cleaned_MBA.csv')
```

```python
new_df = new_df.astype(bool)
```

```python
#Generatig frequent itemsets
my_frequent_itemsets = apriori(new_df, min_support=0.01, use_colnames=True)
```

```python
#generating rules
my_rules = association_rules(my_frequent_itemsets, metric="lift", min_threshold=0.8)
```
